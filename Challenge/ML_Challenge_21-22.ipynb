## 2.1 Clean and Load the Dataset (up to 1 of 11.2 points)
Use the following two cells (a code cell and, a markdown cell) to:

```python
import warnings
warnings.filterwarnings('ignore')

# write here the code for 2.1 Load the Dataset

# In the dataset in line 39537 there were 62 fields instead of 61
# Because there was an additional ',' and i remove that from the csv file
# Also i notice that in line 36263, 39592, 39644 and 39647 were some newlines that causes for example that in "url" got 
# float values instead of a string. So I remove this issue.
# Now we can load the dataset
import pandas as pd
import numpy as np
# We read the dataset
# I use error_bad_lines to detect some bad lines such the 39537 that i mentioned before
df = pd.read_csv('OnlineNewsPopularity.csv', error_bad_lines = False, low_memory=False)

# Now I need to clean the dataset from missing values and some noisy data
# Another problem to handle is the "n.a." because it's a string and the imputer cannot convert a string to a float
# So I transform n.a. in np.nan
index_na = np.where(df.values==' n.a.')
for i in range(len(index_na[0])): 
    df.iloc[index_na[0][i], index_na[1][i]] = np.nan
# I use the Nearest Neighbour Imputation for this reason
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5, weights="uniform")
# Before the transform we got 5 NaN values
print(df.isnull().sum().sum())
df.iloc[:, 1:] = pd.DataFrame(imputer.fit_transform(df.iloc[:, 1:]),columns = df.columns[1:])
# if sum == 0 we don't have NaN values
print(df.isnull().sum().sum())
```

```python
# write here the code for step 3 Model Selection

# 1) I split the dataset after the discretization for the hyperparameter tuning

# 2) I encode nothing

# 3) Discretization of the target feature

# discretization with function qcut of pandas library
y_discretized = pd.qcut(df[' shares'], 5, labels=False)

# scale X
X_scaled = preprocessing.scale(X)

# SMOTE for imbalanced classes
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X_smoted, y_smoted = oversample.fit_resample(X_scaled, y_discretized)

# 4) Now the model to predict the share I decide to test all
from sklearn import svm, tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

# SVM
svc = svm.LinearSVC()

# DecisionTree
dtc = tree.DecisionTreeClassifier()

# RandomForest
rfc = RandomForestClassifier()

# MLPN
mlpc = MLPClassifier()
```

```python
# 5) Hyperparameter tuning

# split dataset for hyperparameter tuning, 50% test 50% training
X_train,X_test,y_train,y_test = train_test_split(X_smoted,y_smoted,test_size=0.5,random_state=0)


# For the first time I perform HyperParameter tuning on the SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# Set the parameters by cross-validation
# Choice of the Kernel, C, gamma for the SVC
tuned_parameters = [{'C': [1, 10, 20, 50, 100],
                    'tol': [1e-4, 1e-5, 1e-7, 0.1]
                    }]

# scores
scores = ['precision', 'recall']
for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(
        svc, tuned_parameters, scoring='%s_macro' % score, n_jobs=-1
    )
    clf.fit(X_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()

    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred, zero_division=0))
    print()
```

```python
# HyperParameter tuning on DTC

# Set the parameters by cross-validation
# Choice of the max_depth, min_samples_leaf and criterion
tuned_parameters = [{'max_depth': [2, 5, 20, 50, 100],
    'min_samples_leaf': [2, 5, 20, 50, 100],
    'criterion': ["gini", "entropy"]}]

# scores
scores = ['precision', 'recall']
for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(
        dtc, tuned_parameters, scoring='%s_macro' % score, n_jobs=-1
    )
    clf.fit(X_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()

    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred, zero_division=0))
    print()
```
